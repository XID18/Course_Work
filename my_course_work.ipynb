Подготовительные операции. Создание датафреймов их их объединение

```python
# Открываем файл courses.csv

import pandas as pd
df1 = pd.read_csv('courses.csv')
df2 = pd.read_csv('course_contents.csv')
df3 = pd.read_csv('progresses.csv')
df4 = pd.read_csv('progress_phases.csv')
df5 = pd.read_csv('students.csv')
# Удаляем столбец Unnamed: 0
df1.drop('Unnamed: 0', axis=1, inplace=True)

# Переименовываем на месте столбец id на course_id
df1.rename(columns={'id': 'course_id'}, inplace=True)

# Объединяем датафреймы courses и course_contents

df21 = pd.merge(df2, df1, left_on=['course_id'], right_on=['course_id'], how='left')

# Объединяем датафреймы progresses и progress_phases
df43 = pd.merge(df4, df3, left_on=['progress_id'], right_on=['id'], how='left')

# Объединяем датафреймы courses, course_contents, progresses с progress_phases
df_main = pd.merge(df43, df21, left_on=['course_id', 'module_number', 'lesson_number'], right_on=['course_id', 'module_number', 'lesson_number'], how='left')

# Удаляем столбец id как дублирующий на месте
df_main.drop(['id'], axis=1, inplace=True)

df_main.head(2)
df_main[df_main.student_id == '768c2987a744c51ce64a5993a2a94eaf'].head(5)
```

```python
# Вычисление минимального, максимального, среднего, медианного возрастов студентов

# Подготовка датафрейма
import datetime
from datetime import datetime, date

df_copy['today'] = date.today() # вставляем столбец текущей даты
df_age = df_copy.dropna() # удаляем строки с отсутствующимим значениями


# df_age['birthday'] =  pd.to_datetime(df_age['birthday'], errors='coerce')
# df_age['today'] =  pd.to_datetime(df_age['today'], errors='coerce')

# Изменяем тип данных со строчного на дату
df_age.loc[:,'birthday'] =  pd.to_datetime(df_age.loc[:,'birthday'], errors='coerce')
df_age.loc[:,'today'] =  pd.to_datetime(df_age.loc[:,'today'], errors='coerce')

# Удаляем строки с отсутствующимим значениями
df_age = df_age.dropna()

# Вычисляем возраст студентов в годах
f = lambda x: x[1] - x[0] # вычисляем разницу между значениями в столбцах: today минус birthday
df_age.loc[:,'Ages'] = df_age.loc[:,['birthday','today']].apply(f, axis=1).astype('<m8[Y]') # данный тип данных astype('<m8[Y]')
                                                                                      # выводит года

# Чистим датафрейм, оставляем возраст для студентов более 18 лет, но менее 60 лет

df_age = df_age[(df_age['Ages'] > 18) & (df_age['Ages'] < 60)]

#Вычисляем минимальный, максимальный, средний, медианный возраст студентов

min = df_age['Ages'].min()
max = df_age['Ages'].max()
mean = df_age['Ages'].mean()
median = df_age['Ages'].median()

# Выводим на печать полученные результаты
print(f'В датафрейме минимальный возраст: {min} лет')
print(f'В датафрейме максимальный возраст: {max} лет')
print(f'В датафрейме средний возраст: {mean} лет')
print(f'В датафрейме медианный возраст: {median} лет')
```

```python
# Задача: вычислить минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

df_age.rename(columns={'id': 'student_id'}, inplace=True) # Переименуем столбец id на student_id для дальнейшего использования

# Объединим два датафрейма
df_union = df_main.merge(df_age, how='outer')

# Рассчитаем минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

df_union_analisys = df_union.groupby("title").agg({'Ages':['min', 'max', 'mean', 'median']})
df_union_analisys.head()
```

```python
# Задача: Постройте bar-chart, отражающий количество студентов на каждом курсе. 
# Ticks нужно развернуть так, чтобы они были читаемы


# Импортируем библиотеку matplotlib
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import register_matplotlib_converters

# используется для вывода графиков здесь же.
%matplotlib inline 

# Подсчитываем студентов по курсам
df_student = df_main.groupby("title").agg({'student_id':'nunique'})
df_student = df_student.reset_index() # Сбрасываем индекс

# используется для нормального отображения индексов pandas на графике
register_matplotlib_converters()

# Для построения графика получаем доступ к объектам Figure и Subplot
fig, subplot = plt.subplots()

# формируем значения по оси X
x_ticks = df_student['title'].values

# формируем значения по оси Y
y_ticks = df_student['student_id'].values

# график функции bar-chart
subplot.bar(x_ticks, y_ticks, color='g', alpha = 1)

# Для того, чтобы Ticks были читаемы, делаем их вертикальными
plt.xticks(x_ticks, rotation='vertical')

# Строим график
plt.show()
```

```python
# Задача: Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, 
# отражающий количество студентов на каждом курсе. График должен иметь заголовок. 
# Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, 
# к которой относится курс (то есть нужна легенда). 
# Прозрачность должна стоять на отметке 0.1. 
# На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐

import numpy as np

# Производим группировку студентов по курсу и направлению. Подсчитываем количество студентов на каждом курсе.
df_student = df_main.groupby(["title", 'field']).agg({'student_id':'nunique'})
df_student = df_student.reset_index() # Сбрасываем индекс
df_student = df_student.sort_values(by='student_id', ascending=False) # Сортируем по возрастанию

# для нормального отображения индексов pandas на графике
register_matplotlib_converters()

# доступ к Figure и Subplot
fig, subplot = plt.subplots()

# формируем значения по оси  X
x_ticks = df_student['title'].values

# формируем значения по оси Y
y_ticks = df_student['student_id'].values

# формируем значения легенды
field = df_student['field'].unique()

# Задаем цвет каждого отдельного бара, указав в параметре clr цвета
clr = ('blue', 'forestgreen', 'gold', 'red')


# Создаем горизонтальный график bar-chart
# Ставим прозрачность: 0.1.
bh = subplot.barh(x_ticks, y_ticks, label=field, color = clr, alpha=0.5)

# Подписываем график
subplot.set_title('Количество студентов на каждом курсе')

# Отображаем легенду на графике
plt.legend(bh, field)

# Рассчитываем медиану
median = df_student['student_id'].median()

# Наносим линию медианы. У медианы цвет - зеленый. Минимум медианы на графике равен 0, максимум - рассчитывается по формуле:

subplot.vlines(median, 0, len(df_student['student_id'].values) - 1, color = 'g')

# Удаляем рамки у графика

subplot.spines['top'].set_visible(False)
subplot.spines['right'].set_visible(False)
subplot.spines['bottom'].set_visible(False)
subplot.spines['left'].set_visible(False)

# Строим график
plt.show()
```

```python
# Выводим датафрейм только по домашним работам
df_main_1 = df_main[(df_main['is_homework']==True)].copy()

# Заменяем значения Nan в столбце status на nodone на месте
# df_main_1.loc[:,'status'].fillna('nodone', inplace = True)

# Агрегируем датафрейм студент - курс: cтатус.
df_student = df_main_1.groupby(["student_id", 'title', 'status']).agg({'status':'count'}).unstack()

# Удаляем иерархический заголовок со словом status
df_student = df_student['status'].reset_index()

# Сортируем по убыванию по столбцу nodone
#df_student = df_student.sort_values('nodone', ascending=False)

# Удаляем ненужные столбцы
df_student.drop(['fail', 'wait'], axis=1, inplace=True)

# Меняем в столбце done и start Nan на ноль
df_student.loc[:, ['done', 'start']] = df_student.loc[:, ['done', 'start']].fillna(0)

# Выводим датафрейм сортировки студент - количество курсов с сортировкой.
df_student_courses = df_main_1.groupby("student_id").agg({'course_id':'nunique'}).sort_values('course_id', ascending=False)

# Сбрасываем индексацию датафрейма
df_student_courses.reset_index(inplace=True)

# Переименовываем столбец
df_student_courses.rename(columns={'course_id': 'count_courses'}, inplace=True)

# Соединил две таблицы: df_student и df_student_courses
df_merge_1 = df_student.merge(df_student_courses)

# Оставляем в датафрейме только строки со столбцом done = 0, start = 0 и количество проходящих курсов больше 2
df_merge_1 = df_merge_1[(df_merge_1['done'] == 0) & (df_merge_1['start'] == 0) & (df_merge_1['count_courses'] > 2)]

# Удаляем ненужные столбцы на месте
df_merge_1.drop(['done', 'start'], axis=1, inplace=True)

# Создаем индекс датафрейма и сортируем датафрейм по нему на месте
df_merge_1 = df_merge_1.set_index('title').sort_values('title', ascending=True)

# Агрегируем датафрейм  по курсам с целью выявления количества 'проблемных' студентов
df_merge_2 = df_merge_1.groupby("title").agg({'student_id':'count'}).sort_values('student_id', ascending=False)

# Расчет количества "проблемных" студентов
sum = df_merge_2['student_id'].sum()

# Расчет доли "проблемных" студентов курса от общего количества "проблемных" студентов
df_merge_2.loc[:, 'percentage'] = df_merge_2.loc[:, 'student_id'] / sum * 100

# Округление два знака после запятой по столбцу "percentage"
df_merge_2.loc[:, 'percentage'] = df_merge_2.loc[:, 'percentage'].round(2)
df_merge_2.head(10)
```
